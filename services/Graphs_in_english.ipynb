```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_processed.csv')
data
```

```python
explode = (0, 0.05)
sizes = [data.GenderCode[data['GenderCode']==0].count(), data.GenderCode[data['GenderCode']==1].count()]
labels = 'Men', 'Women'
figure1, ax1 = plt.subplots(figsize=(5, 5))
colors = ['#2C7873', '#F69454']  
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=colors)
ax1.axis('equal') 
plt.title("Distribution of men and women in the dataset", size=14)
plt.show()
```

```python
male = data[data['GenderCode'] == 0]
female = data[data['GenderCode'] == 1]
bins = np.arange(20, 90, 10)

plt.figure(figsize=(7.5, 5))
plt.hist([male['Age'], female['Age']], bins=bins, color=['#2C7873', '#F69454'], label=['Men', 'Women'], align='left')
plt.xlabel('Age groups')
plt.ylabel('Number of participants')
plt.title('Age of participants')
plt.xticks(bins[:-1], [f'{i}-{i+9}' for i in bins[:-1]])
plt.legend()
plt.grid(True)
plt.show()
```

```python
bins = np.arange(60, 140, 10)
labels = [f'{i}-{i+9}' for i in bins[:-1]]

data['HeartRateGroup'] = np.digitize(data['HeartRate'], bins)
average_calories_by_heartrate = data.groupby(['HeartRateGroup', 'GenderCode'])['Calories'].mean().unstack()

average_calories_by_heartrate.plot(kind='bar', figsize=(16, 7), ax=plt.gca(), color=['#2C7873', '#F69454'])
plt.xlabel('Heart rate')
plt.ylabel('Average number of kilocalories')
plt.title('Relationship between kilocalories burned during exercise and heart rate')
plt.legend(['Men', 'Women'])
plt.grid(True)
plt.xticks(ticks=np.arange(len(labels)), labels=labels, rotation=0)
plt.show()
```

```python
bins = np.arange(37.0, 43.0, 1.0)
labels = [f'{i}-{i+0.9}' for i in bins[:-1]]

data['TemperatureGroup'] = np.digitize(data['BodyTemp'], bins)
average_calories_by_temperature = data.groupby(['TemperatureGroup', 'GenderCode'])['Calories'].mean().unstack()

average_calories_by_temperature.plot(kind='bar', figsize=(16, 7), ax=plt.gca(), color=['#2C7873', '#F69454'])
plt.xlabel('Temperature')
plt.ylabel('Average number of kilocalories')
plt.title('Relationship between kilocalories burned during exercise and body temperature')
plt.legend(['Men', 'Women'], loc='upper left')
plt.grid(True)
plt.xticks(ticks=np.arange(len(labels)), labels=labels, rotation=0)
plt.show()
```

```python
data_filtered = data[data['ExerciseNameUKR'] != 'Полив газону або саду']

people_counts_by_exercise = data_filtered.groupby(['ExerciseNameENG', 'GenderCode']).size().unstack(fill_value=0)

people_counts_by_exercise.plot(kind='bar', figsize=(16, 5), ax=plt.gca(), color=['#2C7873', '#F69454'])
plt.xlabel('Type of exercise')
plt.ylabel('Number of participants')
plt.title('Exercises and the number of participants who performed them')
plt.legend(['Men', 'Women'])
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```
